# Reglas de asociación: caso didáctico mejorado

Este notebook presenta un caso completo, pero todavía fácil de seguir, para aprender **reglas de asociación** en minería de datos.

La idea central es descubrir reglas del tipo:

$$
A \rightarrow B
$$

Por ejemplo:

$$
\{\text{Café}\} \rightarrow \{\text{Pan}\}
$$

Esto se lee como: *cuando una compra contiene café, también suele contener pan*.

## Objetivo del caso

Supongamos que administramos un supermercado pequeño y queremos responder una pregunta práctica:

> ¿Qué productos tienden a comprarse juntos y podrían usarse para diseñar promociones, recomendaciones o decisiones de ubicación en tienda?

Para responder, usaremos:

1. Transacciones de ejemplo.
2. Matriz binaria de productos.
3. Itemsets frecuentes.
4. Reglas de asociación.
5. Métricas: soporte, confianza y lift.
6. Interpretación de reglas buenas, débiles y engañosas.

## 1. Conceptos básicos

Una **transacción** es una compra realizada por un cliente. Cada transacción contiene uno o más productos.

Un **ítem** es un producto individual, por ejemplo: `Pan`, `Café` o `Mantequilla`.

Un **itemset** es un conjunto de ítems, por ejemplo:

$$
\{\text{Pan}, \text{Café}\}
$$

Una **regla de asociación** tiene la forma:

$$
A \rightarrow B
$$

Donde:

- $A$ es el **antecedente**.
- $B$ es el **consecuente**.

Ejemplo:

$$
\{\text{Café}\} \rightarrow \{\text{Pan}\}
$$

Se interpreta como: *si aparece Café en una compra, entonces también podría aparecer Pan*.

## 2. Métricas principales

Para explicar las métricas usaremos el mismo ejemplo durante toda la sección.

Supongamos que tenemos 20 transacciones de un supermercado y queremos analizar la posible relación entre `{Café}` y `{Pan}`.

En el conjunto de datos ocurre lo siguiente:

| Conteo | Valor |
|---|---:|
| Total de transacciones | 20 |
| Transacciones que contienen `{Café}` | 8 |
| Transacciones que contienen `{Pan}` | 12 |
| Transacciones que contienen `{Café, Pan}` | 6 |

Con estos datos analizaremos la regla:

$$
\{\text{Café}\} \rightarrow \{\text{Pan}\}
$$

Esta regla se lee como:

> Si una compra contiene café, entonces también podría contener pan.

### Soporte

El **soporte** mide qué tan frecuente aparece un producto o conjunto de productos en todas las transacciones.

En otras palabras, responde a la pregunta:

> ¿En qué porcentaje de las compras aparece este producto o conjunto de productos?

La fórmula general es:

$$
\text{soporte}(A) = \frac{\text{número de transacciones que contienen } A}{\text{número total de transacciones}}
$$

En nuestro ejemplo hay 20 transacciones en total.

Primero calculamos el soporte de `{Café}`:

$$
\text{soporte}(\{\text{Café}\}) = \frac{8}{20} = 0.40
$$

Esto significa que `{Café}` aparece en el **40%** de las transacciones.

Luego calculamos el soporte de `{Pan}`:

$$
\text{soporte}(\{\text{Pan}\}) = \frac{12}{20} = 0.60
$$

Esto significa que `{Pan}` aparece en el **60%** de las transacciones.

Finalmente, calculamos el soporte conjunto de `{Café, Pan}`:

$$
\text{soporte}(\{\text{Café}, \text{Pan}\}) = \frac{6}{20} = 0.30
$$

Esto significa que `{Café}` y `{Pan}` aparecen juntos en el **30%** de las transacciones.

En términos simples:

> De cada 20 compras, 6 incluyen café y pan juntos.

Este último valor será importante para calcular la confianza de la regla:

$$
\{\text{Café}\} \rightarrow \{\text{Pan}\}
$$

porque la confianza necesita saber cuántas veces aparecen juntos el antecedente `{Café}` y el consecuente `{Pan}`.

### Confianza

La **confianza** mide qué tan seguido aparece el consecuente de una regla cuando aparece su antecedente.

En otras palabras, responde a la pregunta:

> Si una compra contiene `{Café}`, ¿qué tan probable es que también contenga `{Pan}`?

La fórmula general es:

$$
\text{confianza}(A \rightarrow B) = \frac{\text{soporte}(A \cup B)}{\text{soporte}(A)}
$$

También se puede expresar usando conteos:

$$
\text{confianza}(A \rightarrow B) =
\frac{\text{número de transacciones que contienen A y B}}
{\text{número de transacciones que contienen A}}
$$

En nuestro ejemplo queremos analizar la regla:

$$
\{\text{Café}\} \rightarrow \{\text{Pan}\}
$$

Sabemos que:

| Conteo | Valor |
|---|---:|
| Transacciones que contienen `{Café}` | 8 |
| Transacciones que contienen `{Café, Pan}` | 6 |

Entonces:

$$
\text{confianza}(\{\text{Café}\} \rightarrow \{\text{Pan}\}) =
\frac{6}{8} = 0.75
$$

Esto significa que, de las 8 compras que contienen `{Café}`, 6 también contienen `{Pan}`.

En palabras simples:

> Cuando aparece café en una compra, pan también aparece en el 75% de esos casos.

Por eso, la regla:

$$
\{\text{Café}\} \rightarrow \{\text{Pan}\}
$$

tiene una confianza de **0.75**, o **75%**.

Este valor indica que la regla parece bastante fuerte. Sin embargo, todavía falta una pregunta importante:

> ¿Ese 75% es realmente interesante, o simplemente ocurre porque `{Pan}` es un producto muy frecuente?

Para responder eso usamos la siguiente métrica: **lift**.

### Lift

El **lift** mide si la relación entre el antecedente y el consecuente es realmente interesante o si podría explicarse solo porque el consecuente es muy frecuente.

En otras palabras, responde a la pregunta:

> ¿Comprar `{Café}` aumenta la probabilidad de comprar `{Pan}` en comparación con la frecuencia general de `{Pan}`?

La fórmula general es:

$$
\text{lift}(A \rightarrow B) = \frac{\text{confianza}(A \rightarrow B)}{\text{soporte}(B)}
$$

En nuestro ejemplo queremos seguir analizando la regla:

$$
\{\text{Café}\} \rightarrow \{\text{Pan}\}
$$

Ya sabemos que:

| Métrica o conteo | Valor |
|---|---:|
| Total de transacciones | 20 |
| Transacciones que contienen `{Pan}` | 12 |
| Transacciones que contienen `{Café}` | 8 |
| Transacciones que contienen `{Café, Pan}` | 6 |
| Soporte de `{Pan}` | 0.60 |
| Confianza de `{Café} → {Pan}` | 0.75 |

Primero recordamos el soporte de `{Pan}`:

$$
\text{soporte}(\{\text{Pan}\}) = \frac{12}{20} = 0.60
$$

Esto significa que `{Pan}` aparece en el **60%** de todas las compras.

Luego usamos la confianza calculada antes:

$$
\text{confianza}(\{\text{Café}\} \rightarrow \{\text{Pan}\}) = 0.75
$$

Ahora calculamos el lift:

$$
\text{lift}(\{\text{Café}\} \rightarrow \{\text{Pan}\}) =
\frac{0.75}{0.60} = 1.25
$$

El resultado es:

$$
\text{lift} = 1.25
$$

Esto significa que `{Pan}` aparece con mayor frecuencia cuando también aparece `{Café}`.

En términos simples:

> En general, `{Pan}` aparece en el 60% de las compras.  
> Pero cuando una compra contiene `{Café}`, `{Pan}` aparece en el 75% de los casos.

Como 75% es mayor que 60%, la presencia de `{Café}` aumenta la probabilidad de encontrar `{Pan}`.

Por eso, el lift es mayor que 1.

| Lift | Interpretación |
|---:|---|
| $> 1$ | Asociación positiva: A aumenta la aparición de B |
| $= 1$ | Independencia aproximada: A no cambia la aparición de B |
| $< 1$ | Asociación negativa: A reduce la aparición de B |

En este caso:

$$
1.25 > 1
$$

Por lo tanto, la regla:

$$
\{\text{Café}\} \rightarrow \{\text{Pan}\}
$$

muestra una **asociación positiva**.

La interpretación final sería:

> Las compras que contienen café tienen una probabilidad mayor de incluir pan que una compra cualquiera del supermercado.

Sin embargo, el lift de 1.25 no indica una relación extremadamente fuerte. Indica una relación positiva, pero moderada. Esto es importante porque una regla puede tener buena confianza, pero no necesariamente ser muy interesante si el consecuente ya aparece con mucha frecuencia.
